```python
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
import numpy as np
from numbers import Rational
import random
import json
from matplotlib import cm
import math
from matplotlib.patches import Polygon
```

```python
def hue_to_rgb(hue):
    rgb = hsv_to_rgb([hue, 1, 1])
    return "#{:02x}{:02x}{:02x}".format(int(rgb[0]*255), int(rgb[1]*255), int(rgb[2]*255))
```

```python
import uuid
def draw_circles(dpi, size, radius_1, center_x_1, center_y_1, color_1, radius_2, center_x_2, center_y_2, color_2, dis, rot):
    h=5
    fig, ax = plt.subplots(figsize=(h, h),dpi=dpi)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(0, h)
    ax.set_ylim(0, h)
    ax.axis("off")

    circle1_plot = plt.Circle((center_x_1 * h, center_y_1 * h), radius_1 * h, color=hue_to_rgb(color_1), fill=True, linewidth=0.0)
    ax.add_artist(circle1_plot)

    circle2_plot = plt.Circle((center_x_2 * h, center_y_2 * h), radius_2 * h, color=hue_to_rgb(color_2), fill=True, linewidth=0.0)
    ax.add_artist(circle2_plot)

    plt.show()
    ii = str(uuid.uuid4())
    fig.savefig("./2circles-dataset/" + ii + '.png', bbox_inches='tight', dpi=dpi, pad_inches=0)
    plt.show()
    plt.close(fig)
    return ii
```

```python
from numbers import Rational
import random
import json


dpi = [300]
config = ['diagonal_1', 'diagonal_2', 'horizontal', 'vertical']
metadata = dict()

for p in [1, 2]:
  for d in dpi:
    for r in list(np.arange(3,101)):
      rad = 0.5/r
      for c in config:
        for dis in [-.25, -0.2,-0.15, -0.1, -0.05,0.0,0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5,0.55]:
          dis *= (2 * rad)
          if p == 1:
            if c == 'diagonal_1':
              a = np.sqrt(2) * (dis + 2 * rad) / 2
              x1 = 0.5 + (0.5 - a) / 2
              y1 = 0.5 + a + (0.5 - a) / 2
              x2 = x1 + a
              y2 = x1
            elif c == 'diagonal_2':
              a = np.sqrt(2) * (dis + 2 * rad) / 2
              x1 = 0.5 + (0.5 - a) / 2
              y1 = x1
              x2 = x1 + a
              y2 = x2
            elif c == 'horizontal':
              y1 = 0.75
              y2 = 0.75
              x1 = 0.75 - (dis / 2) - rad
              x2 = 0.75 + (dis / 2) + rad
            else:
              x1 = 0.75
              x2 = 0.75
              y1 = 0.75 - (dis / 2) - rad
              y2 = 0.75 + (dis / 2) + rad
          else:
            if c == 'diagonal_1':
              a = np.sqrt(2) * (dis + 2 * rad) / 2
              x1 = (0.5 - a) / 2
              y1 = x1 + a + 0.5
              x2 = x1 + a
              y2 = x1 + 0.5
            elif c == 'diagonal_2':
              a = np.sqrt(2) * (dis + 2 * rad) / 2
              x1 = (0.5 - a) / 2
              y1 = x1 + 0.5
              x2 = x1 + a
              y2 = x2 + 0.5
            elif c == 'horizontal':
              y1 = 0.75
              y2 = 0.75
              x1 = 0.25 - (dis / 2) - rad
              x2 = 0.25 + (dis / 2) + rad
            else:
              x1 = 0.25
              x2 = 0.25
              y1 = 0.75 - (dis / 2) - rad
              y2 = 0.75 + (dis / 2) + rad


          if x1 - rad < 0 or x2 - rad < 0 or y1 - rad < 0 or y2 - rad < 0:
            continue
          if x1 + rad > 1 or x2 + rad > 1 or y1 + rad > 1 or y2 + rad > 1:
            continue

          ii = draw_circles(d, 500, rad, x1, y1, 0.59, rad, x2, y2, 0.82, dis /  (2 * rad), c)

          metadata[ii] = dict()
          metadata[ii]["image_id"] = ii
          metadata[ii]["diameter"] = rad * 2
          metadata[ii]["center_1"] = [x1, y1]
          metadata[ii]["center_2"] = [x2, y2]
          metadata[ii]["distance"] = dis /  (2 * rad)
          metadata[ii]["rotation"] = c
          metadata[ii]["canvas_resolution"] = 1155
          metadata[ii]["patch"] = p
          metadata[ii]["canvas_size"] = 5.0


with open("./2circles-dataset/metadata.json", "w") as fp:
  json.dump(metadata, fp)
```
