```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["GPT-4V", "Gemini Pro", "Claude 3 Sonnet", "Claude 3 Opus"]
words = ["Acknowledgement", "Subdermatoglyphic", "tHyUiKaRbNqWeOpXcZvM"]
prompts = ["P1", "P2"]

data = np.array(
    [
        [[69.17, 68.89], [96.67, 98.33], [82.50, 82.78], [91.39, 90.83]],
        [[63.24, 63.97], [90.69, 91.42], [71.32, 71.57], [94.36, 94.61]],
        [[77.50, 78.33], [86.88, 92.92], [65.83, 66.04], [85.21, 78.96]],
    ]
)

# Set up the plot
fig, axs = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle(
    "Model Performance for Different Prompts in the Circled Word Task", fontsize=16
)

x = np.arange(len(models))
width = 0.35

# Create a grouped bar chart for each word
for i, word in enumerate(words):
    ax = axs[i]
    ax.bar(x - width / 2, data[i, :, 0], width, label="P1", color="skyblue")
    ax.bar(x + width / 2, data[i, :, 1], width, label="P2", color="lightgreen")

    ax.set_ylabel("Accuracy (%)")
    ax.set_title(f"Word: {word}")
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, ha="right")
    ax.legend()
    ax.set_ylim(0, 100)

    # Add value labels on top of each bar
    for j in range(len(models)):
        ax.text(
            j - width / 2,
            data[i, j, 0],
            f"{data[i, j, 0]:.2f}",
            ha="center",
            va="bottom",
        )
        ax.text(
            j + width / 2,
            data[i, j, 1],
            f"{data[i, j, 1]:.2f}",
            ha="center",
            va="bottom",
        )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data from the table
data = {
    "Word": [
        "Acknowledgement",
        "Acknowledgement",
        "Subdermatoglyphic",
        "Subdermatoglyphic",
        "tHyUiKaRbNqWeOpXcZvM",
        "tHyUiKaRbNqWeOpXcZvM",
    ],
    "Font": ["Helvetica", "OpenSans", "Helvetica", "OpenSans", "Helvetica", "OpenSans"],
    "GPT-4V": [69.44, 68.61, 66.42, 60.78, 81.04, 74.79],
    "Gemini Pro": [96.67, 98.33, 90.44, 91.67, 93.12, 86.67],
    "Claude 3 Sonnet": [83.06, 82.22, 68.87, 74.02, 70.00, 61.88],
    "Claude 3 Opus": [90.00, 92.22, 92.89, 96.08, 82.08, 82.08],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Bar width and positions
bar_width = 0.35
models = ["GPT-4V", "Gemini Pro", "Claude 3 Sonnet", "Claude 3 Opus"]
x = np.arange(len(models))
colors = ["#18a181", "#2f96d8", "#cc9c7e", "#d97e61"]
patterns = ["", "//"]  # Empty string for solid, '//' for hatched

for i, word in enumerate(df["Word"].unique()):
    word_data = df[df["Word"] == word]

    helvetica_values = word_data[word_data["Font"] == "Helvetica"][models].values[0]
    opensans_values = word_data[word_data["Font"] == "OpenSans"][models].values[0]

    for j, model in enumerate(models):
        axs[i].bar(
            x[j] - bar_width / 2,
            helvetica_values[j],
            bar_width,
            color=colors[j],
            label=model,
            hatch=patterns[0],
            edgecolor="black",
        )
        axs[i].bar(
            x[j] + bar_width / 2,
            opensans_values[j],
            bar_width,
            color=colors[j],
            hatch=patterns[1],
            edgecolor="black",
        )

    axs[i].set_ylabel("Performance (%)")
    axs[i].set_title(word, fontsize=18)
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(models, rotation=45, ha="right")
    axs[i].set_ylim(0, 100)

    # hide the x tick labels
    axs[i].set_xticklabels([])

    # Add value labels on top of each bar with white background
    for j in range(len(models)):
        axs[i].text(
            j - bar_width / 2,
            helvetica_values[j] - 1.5,
            f"{helvetica_values[j]:.2f}",
            ha="center",
            va="top",
            fontsize=12,
            bbox=dict(facecolor="white", edgecolor="none", alpha=0.7, pad=1),
        )
        axs[i].text(
            j + bar_width / 2,
            opensans_values[j] - 1.5,
            f"{opensans_values[j]:.2f}",
            ha="center",
            va="top",
            fontsize=12,
            bbox=dict(facecolor="white", edgecolor="none", alpha=0.7, pad=1),
        )

    # Create legend for each subplot
    axs[i].legend(title="Models", loc="upper left", bbox_to_anchor=(1, 1))

    # Add font legend
    font_legend_elements = [
        plt.Rectangle(
            (0, 0), 1, 1, facecolor="gray", edgecolor="black", label="Helvetica"
        ),
        plt.Rectangle(
            (0, 0),
            1,
            1,
            facecolor="gray",
            hatch="//",
            edgecolor="black",
            label="OpenSans",
        ),
    ]
    font_legend = axs[i].legend(
        handles=font_legend_elements,
        title="Fonts",
        loc="upper left",
        bbox_to_anchor=(0, 1),
    )

    # Add the font legend manually
    axs[i].add_artist(font_legend)

plt.tight_layout()
plt.savefig(
    "Circled_word_Breakdown_by_font.pdf", bbox_inches="tight", dpi=300, pad_inches=0
)
plt.show()
```
